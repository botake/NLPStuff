Find the nonce such that the string, str(nonce) + “The quick brown fox jumps over the lazy dog” gives a hash with at least 5 trailing zeros. Report the nonce and the first 5 digits of the hash. Explain why we can’t use an optimization algorithm to find such a nonce. 


Using the precept code as a model, a nonce is: 102911, and the first 5 digits of the hash is: d0b8b. Work is shown below: 

We can’t use an optimization algorithm to find such a nonce since the SHA256 cryptographic hash function generates hashes from data that not only cannot be reverse engineered to find the nonce in the data, but also is completely different irrespective of whether the data is similar. In addition, the distribution is so vast that there is no particularly likely answer. Therefore, the best way to find a nonce with a certain number of trailing zeroes is by repeated guessing and checking.

```python
import hashlib
from hashlib import sha256

data = "The quick brown fox jumps over the lazy dog"

MINING_DIFFICULTY = 5

def valid_proof(data, proof):
    guess = f'{data}{proof}'.encode()
    guess_hash = sha256(guess).hexdigest()
    return guess_hash[-MINING_DIFFICULTY:] == '0' * MINING_DIFFICULTY

proof = 0
while valid_proof(data, proof) is False:
    proof += 1
```

Explain why does a consesus algorithm like Proof of Work (PoW) has to include the hash of the previous block. What type of attacks might occur if it didn’t include it?

Proof of Work has to include the hash of the previous block to ensure the validity of the new block being added on to the chain. (The hash of the previous block is what makes a blockchain a “chain” in the first place) In other words, it makes the new incoming block rely on the validity of the previous blocks, which really makes the blockchain immutable, since chainging a old block would require recalculation of all subsequent blocks, which would require enormous computational power. If it didn't include the hash of the previous blocks, it would be vulnerable to all kinds of attacks that take advantage of the fact that old blocks of the blockchain could be modified without impact to the validity of the entire blockchain, since blocks have no way of knowing whether the neighboring blocks are valid or not since they aren't linked. One type of attack include forking attacks.
